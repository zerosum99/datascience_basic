```python
search_word = "투기"  # 검색어 지정
title_list = []
search = f'https://kin.naver.com/search/list.nhn?query={search_word}'
```

```python
def get_titles(search):                  
    
    url = search
    req = requests.get(url)

    if req.ok:
        html = req.text
        soup = BeautifulSoup(html, 'html.parser')
        titles = soup.select(
                'ul.basic1 > li > dl > dt > a'                ## 특정 태그의 정보를 검색한다 
            )
 
        for title in titles:
            title_list.append(title.get_text())  
      
```

```python
def make_word(word_count):                        ##  형태소를 분석해서 리스트에 넣는다. 
    twitter = Okt()
 
    sentences_tag = []
    for sentence in title_list:
        morph = twitter.pos(sentence)
        sentences_tag.append(morph)
 
    noun_adj_list = []
    for sentence1 in sentences_tag:
        for word, tag in sentence1:
            if tag in ['Noun', 'Adjective']:               ## 명사와 형용사만 추출한다. 
                noun_adj_list.append(word)
 
    counts = Counter(noun_adj_list)                         ## 빈도수를 계산한다. 
    tags = counts.most_common(word_count)

    return tags                                             ## 주요 단어만 반환한다 
```

```python
def draw_wordcloud(tags):                                    ## 워드 클라우드를 출력하는 함수를 만든다 
    wc = WordCloud(font_path=path,  mask=r2d2_mask,
                   background_color='white', width=800, height=600, )
    cloud = wc.generate_from_frequencies(dict(tags))
    plt.figure(figsize=(10, 8))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
```
